```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from sklearn.datasets import make_blobs, make_circles, make_moons
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
```

### Linear Kernel and Varied C

```python
X, y = make_blobs(n_samples = 500, n_features = 2, centers = 2, cluster_std = [6, 1], random_state = 10)
```

```python
fig = plt.figure()
plt.scatter([x[0] for x in X], [y[1] for y in X], c = y, cmap = ListedColormap(['red', 'blue']), alpha = 0.6,  edgecolors='black');
plt.title('Linearly separable data')
plt.xlabel('X1')
plt.ylabel('X2')
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 10)
```

```python
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
def plot_decisionboundary(X, y, clf, scaler, margin = 5, title = 'Decision Boundary'):
    inv_X = sc.inverse_transform(X)
    arange_1 = np.arange(start = inv_X[:, 0].min() - margin, stop = inv_X[:, 0].max() + margin, step = 0.1)
    arange_2 = np.arange(start = inv_X[:, 1].min() - margin, stop = inv_X[:, 1].max() + margin, step = 0.1)
    X1, X2 = np.meshgrid(arange_1, arange_2)
    plt.contourf(X1, X2, clf.predict(sc.transform(np.array([X1.ravel(), X2.ravel()]).T)).reshape(X1.shape),
                 cmap = ListedColormap(['tab:red', 'tab:blue']), alpha = 0.3)
    plt.xlim(X1.min(), X1.max())
    plt.ylim(X2.min(), X2.max())
    for i, j in enumerate(np.unique(y)):
        plt.scatter(inv_X[y == j, 0], inv_X[y == j, 1], c = ListedColormap(['red', 'blue'])(i), alpha = 0.6,  edgecolors='black', label = j)
    plt.title(title)
    plt.xlabel('X1')
    plt.ylabel('X2')
    plt.show()
```

```python
try_c = [10**x for x in range(0,6)] 
```

```python
for c in try_c:
    clf = SVC(C = c, kernel = 'linear', random_state = 10)
    clf.fit(X_train, y_train)
    plot_decisionboundary(X_train, y_train, clf, sc, title = f'[Train] Decision Boundary; C = {c}')
    plot_decisionboundary(X_test, y_test, clf, sc, title = f'[Test] Decision Boundary; C = {c}')
```

As the value of C increases the decision boundary becomes less robust. (Very high penalty for miscalssified points)

### Polynomial Kernel

```python
X, y = make_circles(n_samples = 500, factor=0.5, noise=0.02, random_state = 10)
X = 5 * X
```

```python
fig = plt.figure()
plt.scatter([x[0] for x in X], [y[1] for y in X], c = y, cmap = ListedColormap(['red', 'blue']), alpha = 0.6,  edgecolors='black');
plt.title('Linearly inseparable data')
plt.xlabel('X1')
plt.ylabel('X2')
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 10)
```

```python
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
clf_poly = SVC(kernel = 'poly', degree = 2, random_state = 10)
clf_poly.fit(X_train, y_train)
```

```python
plot_decisionboundary(X_train, y_train, clf_poly, sc, title = 'Decision Boundary; Degree 2 Polynomial')
```

```python
plot_decisionboundary(X_test, y_test, clf_poly, sc, title = 'Decision Boundary; Degree 2 Polynomial')
```

### Radial Basis Function Kernel

```python
X, y = make_moons(n_samples = 500, noise=0.03, random_state = 10)
X = 5 * X
```

```python
fig = plt.figure()
plt.scatter([x[0] for x in X], [y[1] for y in X], c = y, cmap = ListedColormap(['red', 'blue']), alpha = 0.6,  edgecolors='black');
plt.title('Linearly inseparable data')
plt.xlabel('X1')
plt.ylabel('X2')
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 10)
```

```python
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
clf_rbf = SVC(kernel = 'rbf', random_state = 10)
clf_rbf.fit(X_train, y_train)
```

```python
plot_decisionboundary(X_train, y_train, clf_rbf, sc, title = 'Decision Boundary; RBF')
```

```python
plot_decisionboundary(X_test, y_test, clf_rbf, sc, title = 'Decision Boundary; RBF')
```
