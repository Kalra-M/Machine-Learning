```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from sklearn.datasets import make_blobs, load_iris, make_circles, make_moons
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier
from itertools import combinations
```

### Varied max_depth

```python
X, y = make_blobs(n_samples = 500, n_features = 2, centers = 3, cluster_std = [5, 3, 2], random_state = 10)
```

```python
fig = plt.figure()
plt.scatter([x[0] for x in X], [y[1] for y in X], c = y, cmap = ListedColormap(['red', 'blue', 'green']), alpha = 0.6,  edgecolors='black');
plt.title('Scatter Plot')
plt.xlabel('X1')
plt.ylabel('X2')
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 10)
```

```python
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
def plot_decisionboundary(X, y, clf, scaler, margin = 5, title = 'Decision Boundary', xlabel = 'X1', ylabel = 'X2'):
    inv_X = sc.inverse_transform(X)
    arange_1 = np.arange(start = inv_X[:, 0].min() - margin, stop = inv_X[:, 0].max() + margin, step = 0.1)
    arange_2 = np.arange(start = inv_X[:, 1].min() - margin, stop = inv_X[:, 1].max() + margin, step = 0.1)
    X1, X2 = np.meshgrid(arange_1, arange_2)
    plt.contourf(X1, X2, clf.predict(sc.transform(np.array([X1.ravel(), X2.ravel()]).T)).reshape(X1.shape),
                 cmap = ListedColormap(['tab:red', 'tab:blue', 'tab:green']), alpha = 0.3)
    plt.xlim(X1.min(), X1.max())
    plt.ylim(X2.min(), X2.max())
    for i, j in enumerate(np.unique(y)):
        plt.scatter(inv_X[y == j, 0], inv_X[y == j, 1], c = ListedColormap(['red', 'blue', 'green'])(i), alpha = 0.6,  edgecolors='black', label = j)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.show()
```

```python
n_depth = [2, 3, 4, 5, None]
```

```python
for depth in n_depth:
    clf = DecisionTreeClassifier(max_depth = depth, random_state = 10)
    clf.fit(X_train, y_train)
    plot_decisionboundary(X_train, y_train, clf, sc, title = f'[Train] Decision Boundaries; Max_depth = {depth}')
    plot_decisionboundary(X_test, y_test, clf, sc, title = f'[Test] Decision Boundaries; Max_depth = {depth}')
```

### Iris

```python
data = load_iris()
X, y = data.data, data.target
```

```python
a = [x for x in range (0,X.shape[1])]
b = a[:-1]
pairs = list(combinations(a, 2))
```

```python
for pair in pairs:
    X_sub = X[:, pair]
    sc = StandardScaler()
    X_sub = sc.fit_transform(X_sub)
    clf = DecisionTreeClassifier(random_state = 10)
    clf.fit(X_sub, y)
    plot_decisionboundary(X_sub, y, clf, sc, margin =2, xlabel = f'{data.feature_names[pair[0]]}', ylabel = f'{data.feature_names[pair[1]]}')
```

### Circles and Moons (Non Linear)

```python
X, y = make_circles(n_samples = 500, factor= 0.5, noise=0.02, random_state = 10)
X = 5 * X
```

```python
fig = plt.figure()
plt.scatter([x[0] for x in X], [y[1] for y in X], c = y, cmap = ListedColormap(['red', 'blue']), alpha = 0.6,  edgecolors='black');
plt.title('Scatter Plot')
plt.xlabel('X1')
plt.ylabel('X2')
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 10)
```

```python
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
clf = DecisionTreeClassifier(random_state = 10)
clf.fit(X_train, y_train)
```

```python
plot_decisionboundary(X_train, y_train, clf, sc)
```

```python
plot_decisionboundary(X_test, y_test, clf, sc)
```
