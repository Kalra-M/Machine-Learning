```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA, KernelPCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis 
```

```python
X = load_wine().data
y = load_wine().target
```

```python
sc = StandardScaler()
X = sc.fit_transform(X)
```

```python
desired_dim = 2
```

```python
print(f'Original dimensions = {X.shape[1]}')
print(f'Original dimensions = {desired_dim}')
```

```python
def scatter_plot(X, y, title, axis):
    plt.scatter([x[0] for x in X], [x[1] for x in X], c = y)
    plt.title(title)
    plt.xlabel(axis[0])
    plt.ylabel(axis[1])
```

```python
pca = PCA(n_components = desired_dim)
X_pca = pca.fit_transform(X)
```

```python
scatter_plot(X_pca, y, 'PCA',  ('PC1', 'PC2'))
```

```python
lda = LinearDiscriminantAnalysis(n_components = desired_dim)
X_lda = lda.fit_transform(X, y)
```

```python
scatter_plot(X_lda, y, 'LDA',  ('LD1', 'LD2'))
```

```python
kpca_poly = KernelPCA(n_components = desired_dim, kernel = 'poly')
X_kpca_poly = kpca_poly.fit_transform(X)
```

```python
scatter_plot(X_kpca_poly, y, 'KERNEL PCA - POLYNOMIAL',  ('PC1', 'PC2'))
```

```python
kpca_rbf = KernelPCA(n_components = desired_dim, kernel = 'rbf')
X_kpca_rbf = kpca_rbf.fit_transform(X)
```

```python
scatter_plot(X_kpca_rbf, y, 'KERNEL PCA - RADIAL BASIS FUNCTION',  ('PC1', 'PC2'))
```
