```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
n_samples = 50
X = np.linspace(0, 5, n_samples)
y = 1.5 * X**3 +  X**2 + 4 * X + 2 * np.random.normal(0, 2.5, np.size(X))
X = X.reshape((n_samples, 1))
```

```python
fig = plt.figure()
plt.plot(X, y, 'o')
fig.suptitle('DATA POINTS')
plt.xlabel('X');
plt.ylabel('Y');
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
```

```python
quadratic = PolynomialFeatures(degree = 2)
X_2 = quadratic.fit_transform(X_train)[:,1:]
#cubic = PolynomialFeatures(degree = 3)
#X_3 = cubic.fit_transform(X_train)[:,1:]
```

```python
linear_reg, quadratic_reg = LinearRegression(), LinearRegression()
#cubic_reg = LinearRegression()
```

```python
linear_reg.fit(X_train, y_train)
quadratic_reg.fit(X_2, y_train)
#cubic_reg.fit(X_3, y_train)
```

```python
yh1 = linear_reg.predict(X_test)
yh2 = quadratic_reg.predict(quadratic.transform(X_test)[:,1:])
#yh3 = cubic_reg.predict(cubic.transform(X_test)[:,1:])
```

```python
fig = plt.figure()
plt.plot(X_test,y_test,'o', label = 'Data Points')
plt.plot(X_test, yh1, label = 'Linear Fit')
plt.plot(sorted(X_test), sorted(yh2), label = 'Quadratic Fit')
#plt.plot(sorted(X_test), sorted(yh3), label = 'Cubic Fit')
plt.title(f'R Squared : Linear = {round(r2_score(y_test, yh1),2)}, Quadratic = {round(r2_score(y_test, yh2),2)}')
plt.legend()
plt.xlabel('X');
plt.ylabel('Y');
```
