```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
```

```python
df = pd.read_csv('Market_Basket_Optimisation.csv', header = None)
```

```python
df.head()
```

```python
all_transactions = []

for i in range(df.shape[0]):
    all_transactions.append([str(df.values[i, j]) for j in range(df.shape[1])])
```

```python
t_encoder = TransactionEncoder()
encoded = t_encoder.fit_transform(all_transactions)
te_df = pd.DataFrame(encoded, columns = t_encoder.columns_)
```

```python
te_df.head()
```

```python
te_df.drop('nan', axis = 1, inplace = True)
```

```python
rules = apriori(te_df, min_support = 0.02, max_len = 2, use_colnames = True)
```

```python
rules['length'] = rules['itemsets'].apply(lambda x : len(x))
```

```python
rules.head()
```

```python
rules.drop(rules[rules['length'] == 1].index, axis = 0, inplace = True)
```

```python
rules['LHS'] = rules['itemsets'].apply(lambda x : tuple(x)[0])
rules['RHS'] = rules['itemsets'].apply(lambda x : tuple(x)[1])
```

```python
rules.drop(['itemsets', 'length'], axis = 1, inplace = True)
```

```python
rules.reset_index(drop = True, inplace =True)
```
